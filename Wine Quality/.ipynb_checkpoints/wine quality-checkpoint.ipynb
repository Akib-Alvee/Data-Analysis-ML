# Task 02: Exploring the Data

Visualizing the dataset to see the distributions of the predictors of the the dataset.

```python
fig = plt.figure(figsize = (20,20))
fig.suptitle('Histogram of Different Features\n', fontsize=30)
columns=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']

for i in range (12):
    plt.subplot(4,3,i+1)
    sns.distplot(df[columns[i]],hist_kws={'color':'Teal'},kde_kws={'color':'Navy'})
    plt.axvline(df[columns[i]].mean(), color='Black', linestyle='dashed', linewidth=2)
    plt.axvline(df[columns[i]].median(), color='Black', linestyle='-', linewidth=2)
    plt.legend(['kde','mean','median'])
    plt.title('Variation of '+columns[i])
fig=plt.gcf()
fig.set_size_inches(18,18)
fig.tight_layout()
```

We can see some of the predictors are skewed. This is not good for modeling thus, we will do a boxcox transformation in order to remove the skewness in the dataset.But first we need check the skewness of the data.

```python
def skewness_check(data):
    skew_value = list(st.skew(data))
    skew_string = []

    for skew in skew_value:
        if skew >= -.5 and skew <= .5:
            skew_string.append("Light Skewed")
        elif skew <= -.5 and skew >= -1 and skew <= .5 and skew >= 1:
            skew_string.append("Moderately Skewed")
        else:
            skew_string.append("Heavily Skewed")
    skew_df = pd.DataFrame({'Column': data.columns, 'Skewness': skew_value, 'Skew Category': skew_string})
    return skew_df

skewness_check(df.iloc[:, :-1])
```

Here also we have a lot of heavily skewed values. So applying boxcox transformation should considerably helps with removing the skewness.We have to move the whole distribution to the right a bit for citric acid to do the boxcox transformation.

```python
def box_plot(data, height = 10, width = 15, title = "Boxplot"):
    plt.figure(figsize=(width, height))
    plt.suptitle(title, y = 0.94, fontsize = 30)
    sns.set(font_scale = 2)
    sns.boxplot(data = data, orient = "h")

box_plot(data = df.iloc[:, :-1], title = "With Outliers")
```

```python
df[["citric acid"]] += 0.1
```

```python
# Boxcox Tranformation
def boxcox_trans(data):
    for i in range(data.shape[1]):
        data.iloc[:, i], _ = st.boxcox(data.iloc[:, i])
    return data

# Subset the predcitors
df_trans = df[df>0]
df_trans.iloc[:, :-1] = boxcox_trans(df_trans.iloc[:, :-1])
skewness_check(df_trans.iloc[:, :-1])
```

If we try to see our data after removing the outliers.

```python
box_plot(data = df_trans, title = "After using Boxcox")
```

## Separating Training and Test Data

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split 

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) 
```

# Task 3 : Data Modeling

Here, we are predicting a numerical continuous value,this will be trained on various Regression Models. Regression analysis is a sub field of Supervised Learning.

Coming to training the model for Wine Quality Prediction we will be first pre-processing the data and then running 3 different models on the dataset to single out the best result. The models are as follows:

<ul>
    <li>Multiple Linear Regression</li>
    <li>Lasso Regression</li>
    <li>Support Vector Regression(SVR)</li>
</ul>


### Multiple Linear Regression<br>
If we have one dependent feature and multiple independent features then basically call it a multiple linear regression.Here we have one dependent feature quality,rest of the features are independent. 

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

Error_dict={}

regressor = LinearRegression()
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
plt.scatter(y_test,y_pred)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.show()
```

```python
a = regressor.coef_
b = regressor.intercept_
print("Estimated model slope, a:" , a)
print("Estimated model intercept, b:" , b) 
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("RMSE value: {:.4f}".format(rmse))
```

```python
Error_dict['multiple_linear_regression_r_squared_value']=r2_score(y_test,y_pred)
print ("R2 Score value: {:.4f}".format(r2_score(y_test, y_pred)))
```

```python
accuracy = regressor.score(x_test, y_test)
"Accuracy: {}%".format(int(round(accuracy * 100)))
```

### Lasso Regression<br>
Lasso regression is a regularization technique. It is used over regression methods for a more accurate prediction. This model uses shrinkage. Shrinkage is where data values are shrunk towards a central point as the mean. The lasso procedure encourages simple, sparse models (i.e. models with fewer parameters). This particular type of regression is well-suited for models showing high levels of multicollinearity or when you want to automate certain parts of model selection, like variable selection/parameter elimination.

```python
lasso_regressor= linear_model.Lasso(alpha=0.1)
lasso_regressor.fit(x_train,y_train)

y_pred=lasso_regressor.predict(x_test)

plt.scatter(y_test,y_pred)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
```

```python
a = lasso_regressor.coef_
b = lasso_regressor.intercept_
print("Estimated model slope, a:" , a)
print("Estimated model intercept, b:" , b) 
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("RMSE value: {:.4f}".format(rmse))
```

```python
Error_dict['Lasso_r_squared_value']=r2_score(y_test,y_pred)
print ("R2 Score value: {:.4f}".format(r2_score(y_test, y_pred)))
```

```python
accuracy = lasso_regressor.score(x_test, y_test)
"Accuracy: {}%".format(int(round(accuracy * 100)))
```

## Support Vector Regression(using Kernels)<br>
Support Vector Regression is a supervised learning algorithm that is used to predict discrete values. Support Vector Regression uses the same principle as the SVMs. The basic idea behind SVR is to find the best fit line. In SVR, the best fit line is the hyperplane that has the maximum number of points.

```python
from sklearn.svm import SVR 

svm_linear = SVR(kernel='linear')

svm_linear.fit(x_train,y_train)

y_pred=svm_linear.predict(x_test)

Error_dict['svm_linear_r_squared_value']=r2_score(y_test,y_pred)
```
